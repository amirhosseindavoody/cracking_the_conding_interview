```python
import math

class Node:
    def __init__(self,value,index):
        self.value = value
        self.index = index
    def __repr__(self):
        return "(%s,%s)" % (self.value, self.index)

def merge_sort(arr,aux,left_index,right_index):
    if (right_index-left_index)<=1:
        return
    
    middle_index = math.floor((right_index+left_index)/2.0)
    merge_sort(arr,aux,left_index,middle_index)
    merge_sort(arr,aux,middle_index,right_index)
    
    i = left_index
    j = middle_index
    k = left_index
    while i < middle_index and j < right_index:
        if arr[i].value > arr[j].value:
            aux[k] = arr[j]
            j+=1
        else:
            aux[k] = arr[i]
            i+=1
        k = k+1
    if i < middle_index:
        aux[k:right_index] = arr[i:middle_index]
    if j < right_index:
        aux[k:right_index] = arr[j:right_index]
    arr[left_index:right_index] = aux[left_index:right_index]
    
    return

def binary_search(arr,left,right,value):
    if (arr[left].value == value):
        return True, left
    if (arr[left].value > value):
        return False, 0
    if (arr[right-1].value < value):
        return False, 0
    if (left==right):
        return False, 0
    middle = math.floor((left+right)/2.0)
    if (arr[middle].value <= value):
        return binary_search(arr,middle,right,value)
    if (arr[middle].value > value):
        return binary_search(arr,left,middle,value)

def find_pairs(total,arr):
    for i in range(len(arr)):
        value = total - arr[i].value
        found, j = binary_search(arr,i+1,len(arr),value)
        if found:
            return arr[i].index, arr[j].index
    return False

contents = []
with open('binary_search.in') as fp:
    for line in fp:
        contents.append(line)

t = int(contents[0].strip())
for a0 in range(t):
    m = int(contents[a0*3+1].strip())
    n = int(contents[a0*3+2].strip())
    a = list(map(int, contents[a0*3+3].strip().split(' ')))
    node_list = []
    aux_list = []
    for i in range(n):
        node_list.append(Node(a[i],i+1))
        aux_list.append(Node(0,0))
    merge_sort(node_list,aux_list,0,n)
    print(min(find_pairs(m,node_list)), max(find_pairs(m,node_list)))
```

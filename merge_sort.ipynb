# Merge Sort: Counting Inversions

````python
import math
````

### A simplified version that is not optimized for performace

````python
def count_inversions(a):
    a_sorted = a
    return merge_sort(a)

def merge_sort(arr):
    if len(arr)<=1:
        return 0
    middle_index = math.floor(len(arr)/2.0)
    left_sorted = arr[:middle_index]
    right_sorted = arr[middle_index:]
    n = merge_sort(left_sorted)
    m = merge_sort(right_sorted)
    
    i = 0
    j = 0
    k = 0
    p = n+m
    while i < len(left_sorted) and j < len(right_sorted):
        if left_sorted[i] > right_sorted[j]:
            arr[k] = right_sorted[j]
            j+=1
            p+=len(left_sorted)-i
        else:
            arr[k] = left_sorted[i]
            i+=1
        k = k+1

    if i < len(left_sorted):
        arr[k:] = left_sorted[i:]

    if j < len(right_sorted):
        arr[k:] = right_sorted[j:]
    
    return p
````

### This version is a little bit more optimized

````python
def count_inversions(a):
    helper = len(a)*[0]
    left_index = 0
    right_index = len(a)
    return merge_sort(a,helper,left_index,right_index)

def merge_sort(arr,helper,left_index,right_index):
    if (right_index-left_index)<=1:
        return 0
    
    middle_index = math.floor((right_index+left_index)/2.0)
    n = merge_sort(arr,helper,left_index,middle_index)
    m = merge_sort(arr,helper,middle_index,right_index)
    
    i = left_index
    j = middle_index
    k = left_index
    p = n+m
    while i < middle_index and j < right_index:
        if arr[i] > arr[j]:
            helper[k] = arr[j]
            j+=1
            p+=middle_index-i
        else:
            helper[k] = arr[i]
            i+=1
        k = k+1

    if i < middle_index:
        helper[k:right_index] = arr[i:middle_index]

    if j < right_index:
        helper[k:right_index] = arr[j:right_index]
    
    arr[left_index:right_index] = helper[left_index:right_index]
    
    return p
    
````
